> STEP 3: Data cleaning and preparation

### Subtask:
Handle missing values and potentially convert the `price` column to a numerical format for easier analysis.


```python
print("Missing values per column:")
print(df.isnull().sum())
```

Remove the '$' sign from the `price` column and convert it to a numerical data type, converting errors to NaN.



```python
df['price'] = df['price'].astype(str).str.replace('$', '', regex=False)
df['price'] = pd.to_numeric(df['price'], errors='coerce')

print("\nData types after price conversion:")
print(df.dtypes)

print("\nMissing values in price column after conversion:")
print(df['price'].isnull().sum())
```


Handle missing values in the `price` column by imputing with the median, and handle missing values in other columns with a significant number of missing values by imputing with a placeholder 'Unknown'.



```python
# Impute missing price values with the median
median_price = df['price'].median()
df['price'].fillna(median_price, inplace=True)

# Impute missing values in other columns with 'Unknown'
columns_to_impute_unknown = ['description', 'manufacturer', 'country_of_origin', 'material', 'color', 'package_dimensions']
for col in columns_to_impute_unknown:
    df[col].fillna('Unknown', inplace=True)

print("\nMissing values after handling:")
print(df.isnull().sum())
```

## Visual analysis

### Subtask:
Create charts to visualize distributions and relationships between different features, such as price distribution, top brands, and category breakdowns.


```python
# 1. Histogram of price distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=30, kde=True)
plt.title('Distribution of Product Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python

# 2. Bar plot of top 10 most frequent brands
plt.figure(figsize=(12, 6))
top_brands = df['brand'].value_counts().nlargest(10)
sns.barplot(x=top_brands.index, y=top_brands.values, palette='viridis')
plt.title('Top 10 Most Frequent Brands')
plt.xlabel('Brand')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python

# 3. Bar plot of top 10 most frequent categories
plt.figure(figsize=(12, 6))
# Need to handle the list format in categories column
categories_list = df['categories'].str.strip('[]').str.split(', ')
all_categories = [category.strip("'") for sublist in categories_list.dropna() for category in sublist]
top_categories = pd.Series(all_categories).value_counts().nlargest(10)
sns.barplot(x=top_categories.index, y=top_categories.values, palette='viridis')
plt.title('Top 10 Most Frequent Categories')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

```

```python
# 4. Scatter plot of price against top brands (using a subset for clarity)
plt.figure(figsize=(14, 7))
# Filter for top 10 brands to make the scatter plot readable
df_top_brands = df[df['brand'].isin(top_brands.index)]
sns.swarmplot(x='brand', y='price', data=df_top_brands)
plt.title('Price vs. Top 10 Brands')
plt.xlabel('Brand')
plt.ylabel('Price')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Calculate the count of products for each country of origin
country_counts = df['country_of_origin'].value_counts()

# Select the top countries for the pie chart (you can adjust the number as needed)
# Let's include 'Unknown' as a category since it's a significant portion
top_countries_for_pie = country_counts.nlargest(5) # Adjust number for clarity

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_countries_for_pie, labels=top_countries_for_pie.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Products by Country of Origin')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```



### Subtask:
Explore relationships between price and categories, or analyze the distribution of products by country of origin.


```python
# Preprocess the 'categories' column to extract individual categories and create a new dataframe
categories_list = df['categories'].str.strip('[]').str.split(', ')
category_price_list = []
for index, row in df.iterrows():
    if pd.notna(row['categories']):
        for category in categories_list[index]:
            category_price_list.append({'category': category.strip("'"), 'price': row['price']})

df_categories_price = pd.DataFrame(category_price_list)
```

```python

# Get the top 10 most frequent categories from the new dataframe
top_categories = df_categories_price['category'].value_counts().nlargest(10).index

# Create a box plot to visualize the distribution of prices across the top 10 most frequent categories
plt.figure(figsize=(14, 7))
sns.boxplot(x='category', y='price', data=df_categories_price[df_categories_price['category'].isin(top_categories)])
plt.title('Price Distribution Across Top 10 Categories')
plt.xlabel('Category')
plt.ylabel('Price')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python

# Calculate and display the count of products for each country of origin.
country_counts = df['country_of_origin'].value_counts()
print("\nProduct count by Country of Origin:")
display(country_counts)

# Create a bar plot to visualize the distribution of products by country of origin
plt.figure(figsize=(12, 6))
# Focus on top countries if necessary for clarity
top_countries = country_counts.nlargest(10) # Adjust number for clarity if needed
sns.barplot(x=top_countries.index, y=top_countries.values, palette='viridis',)
plt.title('Distribution of Products by Country of Origin (Top 10)')
plt.xlabel('Country of Origin')
plt.ylabel('Number of Products')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
